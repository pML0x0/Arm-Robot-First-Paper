```python

def ploting_avarage_with_min_max(data, ylabel, xlabel, legendSet, title=None, SaveAs=None):
    
    plt.figure(figsize=(10, 6))  # Example size, adjust as needed
    
    for d, l in zip(data, legendSet):
        avg = np.mean(d, axis=0)
        min_y = np.min(d, axis=0)
        max_y = np.max(d, axis=0)
        x = np.arange(len(avg))

        plt.plot(avg, label=l)
        plt.fill_between(x, min_y, max_y, alpha=0.2, label='_nolegend_')
        
    if title:
        plt.title(title, fontsize=20)
    plt.xticks(fontsize=22)
    plt.yticks(fontsize=22)
    plt.xlabel(xlabel, fontsize=22)
    plt.ylabel(ylabel, fontsize=22)
    plt.legend(loc="upper left", fontsize=18)
    
    plt.tight_layout()  # Adjust layout
    
    if SaveAs:
        plt.savefig(SaveAs, dpi=600, bbox_inches='tight')  # Save with adjustments

# Example usage with your data
```

```python
def ploting_avarage_with_min_max2(data_reward, data_pohibitive_state, ylabel1, ylabel2, xlabel, legendSet1, legendSet2, title = None, SaveAs=None):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))  # Creates two subplots side by side

    # Plotting the first dataset on the first subplot (ax1)
    for d, l in zip(data_reward, legendSet1):
        avg = np.mean(d, axis=0)
        min_y = np.min(d, axis=0)
        max_y = np.max(d, axis=0)
        x = np.arange(len(avg))
        ax1.fill_between(x, min_y, max_y, alpha=0.2, label='_nolegend_')
        ax1.plot(x, avg, label=l)

    # Plotting the second dataset on the second subplot (ax2)
    for d, l in zip(data_pohibitive_state, legendSet2):
        avg = np.mean(d, axis=0)
        min_y = np.min(d, axis=0)
        max_y = np.max(d, axis=0)
        x = np.arange(len(avg))
        ax2.fill_between(x, min_y, max_y, alpha=0.2, label='_nolegend_')
        ax2.plot(x, avg, label=l)

    # Setting titles, labels, and legends for each subplot
    #ax1.set_title(title + " - Reward")
    ax1.set_xlabel(xlabel, fontsize=20)
    ax1.set_ylabel(ylabel1, fontsize=20)
    ax1.legend(loc="upper left")

    #ax2.set_title(title + " - Prohibitive State")
    ax2.set_xlabel(xlabel, fontsize=20)
    ax2.set_ylabel(ylabel2, fontsize=20)
    ax2.legend(loc="upper left", fontsize=20)

    # Adjust the layout so labels do not overlap
    plt.tight_layout()

    # Saving the figure
    if SaveAs:
        fig.savefig(SaveAs, dpi=300)

    # Show the plot
    plt.show()

# Example usage of the function
# ploting_avarage_with_min_max2(data1, data2, "Y Label", "X Label", legend1, legend2, "Title", "filename.png")
```

```python
import matplotlib.pyplot as plt

def plot_bar_graph(data, data2, categories, xlabel, ylabel, legendSet, title=None, SaveAs=None):
    # Set positions of bars on X axis
    barWidth = 0.3
    r1 = range(len(categories))
    r2 = [x + barWidth for x in r1]

    # Create the bar plots side by side
    plt.figure(figsize=(10, 6))
    plt.bar(r1, data, color='blue', width=barWidth, edgecolor='black', label=legendSet[0])
    plt.bar(r2, data2, color='red', width=barWidth, edgecolor='black', label=legendSet[1])

    # Add labels, title, and legend with increased font size
    plt.xlabel(xlabel, fontsize=20)  # Increase font size for x-axis label
    plt.ylabel(ylabel, fontsize=20)  # Increase font size for y-axis label
    plt.xticks(fontsize=12)
    plt.xticks([r + barWidth / 2 for r in range(len(categories))], categories, fontsize=20)  # Increase font size for x-ticks
    plt.title(title, fontsize=20)  # Increase font size for title
    plt.legend(fontsize=20)  # Increase font size for legend

    if SaveAs:
        plt.savefig(SaveAs, dpi=300)

    # Show the plot
    plt.show()
```
